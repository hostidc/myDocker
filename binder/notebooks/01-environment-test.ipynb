# FastGPT 环境测试

这个 Notebook 用于测试 MyBinder 环境中 FastGPT 的各个组件是否正常工作。

## 1. Python 环境检查

```python
import sys
print(f"Python 版本: {sys.version}")
print(f"Python 路径: {sys.executable}")
```

## 2. 检查已安装的包

```python
import pkg_resources

required_packages = [
    'jupyterlab',
    'notebook',
    'fastapi',
    'uvicorn',
    'pymongo',
    'psycopg2',
    'redis',
    'minio',
    'chromadb'
]

for package in required_packages:
    try:
        version = pkg_resources.get_distribution(package).version
        print(f"✓ {package}: {version}")
    except pkg_resources.DistributionNotFound:
        print(f"✗ {package}: 未安装")
```

## 3. 文件系统测试

```python
import os

# 检查工作目录
work_dir = '/home/jovyan/work'
print(f"工作目录: {work_dir}")
print(f"目录存在: {os.path.exists(work_dir)}")
print(f"可写: {os.access(work_dir, os.W_OK)}")

# 创建测试文件
test_file = os.path.join(work_dir, 'test.txt')
with open(test_file, 'w') as f:
    f.write('Hello from FastGPT on Binder!')

print(f"✓ 测试文件已创建: {test_file}")

# 读取测试文件
with open(test_file, 'r') as f:
    content = f.read()
    print(f"文件内容: {content}")
```

## 4. 网络测试

```python
import socket

ports_to_check = [
    (80, 'Nginx'),
    (3000, 'FastGPT App'),
    (8000, 'Backend API'),
    (27017, 'MongoDB'),
    (5432, 'PostgreSQL'),
    (6379, 'Redis'),
    (9000, 'MinIO API'),
    (9001, 'MinIO Console')
]

for port, service in ports_to_check:
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    result = sock.connect_ex(('localhost', port))
    if result == 0:
        print(f"✓ {service} (端口 {port}): 可访问")
    else:
        print(f"✗ {service} (端口 {port}): 无法访问")
    sock.close()
```

## 5. 环境变量检查

```python
import os

binder_vars = [
    'BINDER_LAUNCH_URL',
    'JUPYTERHUB_API_TOKEN',
    'HOME',
    'USER'
]

for var in binder_vars:
    value = os.environ.get(var, '未设置')
    print(f"{var}: {value}")
```
